```python
# 1) Instalación
!pip -q install yfinance requests pandas numpy pyarrow

# 2) Imports y configuración
import time
import math
import warnings
import requests
import numpy as np
import pandas as pd
import yfinance as yf

warnings.filterwarnings("ignore")

START = "2009-01-01"
END = None                 # o "2026-04-19"
PRICE_FIELD = "Close"  # o "Close"
CHUNK_SIZE_QUOTES = 150
CHUNK_SIZE_PRICES = 80
REQUEST_TIMEOUT = 20
PAUSE_BETWEEN_CALLS = 0.15

NAME_TO_TICKER = STOXX600_TICKERS.copy()
TICKERS = list(NAME_TO_TICKER.values())

# Divisas cruce directo contra EUR en Yahoo
FX_TO_EUR = {
    "GBP": "GBPEUR=X",
    "CHF": "CHFEUR=X",
    "SEK": "SEKEUR=X",
    "DKK": "DKKEUR=X",
    "NOK": "NOKEUR=X",
    "PLN": "PLNEUR=X",
    "USD": "USDEUR=X",
    "CZK": "CZKEUR=X",
    "HUF": "HUFEUR=X",
    "RON": "RONEUR=X",
    "ZAR": "ZAREUR=X",
}

# Último recurso manual
MANUAL_CURRENCY_OVERRIDES = {
    "PUIG.MC":"EUR"
    # "SHEL.L": "GBp",
}

session = requests.Session()
session.headers.update({
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json,text/plain,*/*",
})


# 3) Normalizar divisa
def normalize_quote_currency(raw_currency: str):
    """
    Devuelve:
      currency    -> divisa económica real
      unit_factor -> factor para llevar el precio a unidades estándar
    Ejemplo:
      GBp -> GBP y factor 0.01
    """
    if raw_currency is None or pd.isna(raw_currency):
        return None, np.nan

    raw = str(raw_currency).strip()

    special = {
        "GBp": ("GBP", 0.01),
        "GBX": ("GBP", 0.01),
        "GBx": ("GBP", 0.01),
        "ZAc": ("ZAR", 0.01),
    }

    return special.get(raw, (raw, 1.0))


def chunked(seq, size):
    for i in range(0, len(seq), size):
        yield seq[i:i+size]


def fetch_quote_currencies_bulk(symbols, chunk_size=CHUNK_SIZE_QUOTES, timeout=REQUEST_TIMEOUT):
    """
    Resuelve divisas en bloque desde Yahoo quote endpoint.
    Mucho más estable que get_history_metadata() para este uso.
    """
    out = {}

    for chunk in chunked(symbols, chunk_size):
        url = "https://query1.finance.yahoo.com/v7/finance/quote"
        params = {"symbols": ",".join(chunk)}

        try:
            r = session.get(url, params=params, timeout=timeout)
            r.raise_for_status()
            data = r.json()

            results = data.get("quoteResponse", {}).get("result", [])
            for item in results:
                sym = item.get("symbol")
                cur = (
                    item.get("currency")
                    or item.get("financialCurrency")
                    or item.get("currencyCode")
                )
                if sym:
                    out[sym] = cur
        except Exception as e:
            print(f"Fallo en bloque {chunk[:3]}...: {e}")

        print(f"Divisas resueltas por quote bulk: {len(out)}/{len(symbols)}")
        time.sleep(PAUSE_BETWEEN_CALLS)

    return out


def fetch_chart_currency(symbol, timeout=REQUEST_TIMEOUT):
    """
    Fallback por ticker usando el endpoint chart de Yahoo.
    """
    url = f"https://query1.finance.yahoo.com/v8/finance/chart/{symbol}"
    params = {
        "range": "5d",
        "interval": "1d",
        "includePrePost": "false",
        "events": "div,splits,capitalGains",
    }

    try:
        r = session.get(url, params=params, timeout=timeout)
        r.raise_for_status()
        data = r.json()
        meta = data.get("chart", {}).get("result", [{}])[0].get("meta", {})
        return meta.get("currency")
    except Exception:
        return None


def resolve_all_currencies(symbols):
    """
    1) quote bulk
    2) chart fallback
    3) manual overrides
    """
    out = fetch_quote_currencies_bulk(symbols)

    missing = [s for s in symbols if not out.get(s)]
    if missing:
        print(f"\nFallback chart para {len(missing)} tickers...")
        for i, sym in enumerate(missing, 1):
            cur = fetch_chart_currency(sym)
            out[sym] = cur
            if i % 25 == 0 or i == len(missing):
                print(f"Fallback chart: {i}/{len(missing)}")
            time.sleep(0.05)

    # overrides manuales al final
    out.update(MANUAL_CURRENCY_OVERRIDES)

    return out


# 4) Descargar precios
def download_field_in_chunks(tickers, field="Close", start=None, end=None, chunk_size=CHUNK_SIZE_PRICES):
    pieces = []

    for i, chunk in enumerate(chunked(tickers, chunk_size), 1):
        try:
            df = yf.download(
                tickers=chunk,
                start=start,
                end=end,
                auto_adjust=True,
                actions=False,
                repair=True,
                threads=True,
                progress=False,
                group_by="column",
            )
        except TypeError:
            df = yf.download(
                tickers=chunk,
                start=start,
                end=end,
                auto_adjust=True,
                actions=False,
                threads=True,
                progress=False,
                group_by="column",
            )

        if df.empty:
            piece = pd.DataFrame(index=pd.DatetimeIndex([]), columns=chunk)
        elif isinstance(df.columns, pd.MultiIndex):
            level0 = set(df.columns.get_level_values(0))
            chosen = field if field in level0 else "Close"
            piece = df[chosen].copy()
        else:
            chosen = field if field in df.columns else "Close"
            piece = df[[chosen]].copy()
            piece.columns = chunk[:1]

        pieces.append(piece)
        print(f"Precios descargados: {min(i*chunk_size, len(tickers))}/{len(tickers)}")
        time.sleep(PAUSE_BETWEEN_CALLS)

    result = pd.concat(pieces, axis=1)
    result = result.loc[:, ~result.columns.duplicated(keep="first")]
    result = result.sort_index()
    result = result.reindex(columns=tickers)

    return result


# 5) Resolver divisas

raw_currency_map = resolve_all_currencies(TICKERS)

currency_rows = []
for name, ticker in NAME_TO_TICKER.items():
    raw_cur = raw_currency_map.get(ticker)
    currency, unit_factor = normalize_quote_currency(raw_cur)
    currency_rows.append({
        "name": name,
        "ticker": ticker,
        "quote_currency_raw": raw_cur,
        "currency": currency,
        "unit_factor": unit_factor,
    })

currency_report = pd.DataFrame(currency_rows).sort_values(["currency", "ticker"]).reset_index(drop=True)

unresolved = currency_report[currency_report["currency"].isna()].copy()
print("\nTickers sin divisa resuelta:", len(unresolved))
if not unresolved.empty:
    display(unresolved)

    # Si queda alguno, rellenar aquí manualmente y relanzar desde el bloque 5:
    # MANUAL_CURRENCY_OVERRIDES.update({
    #     "XXX": "EUR",
    # })

    raise RuntimeError(
        "Aún quedan tickers sin divisa. Mira unresolved y añade solo esos casos a MANUAL_CURRENCY_OVERRIDES."
    )

print("\nDistribución de divisas detectadas:")
print(currency_report["currency"].value_counts(dropna=False).sort_index())

# 6) Descargar precios nativos
prices_native = download_field_in_chunks(
    TICKERS,
    field=PRICE_FIELD,
    start=START,
    end=END,
    chunk_size=CHUNK_SIZE_PRICES
)

all_nan_cols = prices_native.columns[prices_native.isna().all()].tolist()
if all_nan_cols:
    print("\nTickers sin precios descargados:")
    print(all_nan_cols)

# 7) Descargar FX necesarios contra EUR

curr_by_ticker = currency_report.set_index("ticker")["currency"]
unit_by_ticker = currency_report.set_index("ticker")["unit_factor"].astype(float)

needed_fx_ccy = sorted(set(curr_by_ticker.dropna()) - {"EUR"})
missing_fx = [c for c in needed_fx_ccy if c not in FX_TO_EUR]

if missing_fx:
    raise RuntimeError(
        f"Faltan pares FX contra EUR para estas divisas: {missing_fx}. Añádelas a FX_TO_EUR."
    )

fx_symbols = [FX_TO_EUR[c] for c in needed_fx_ccy]

if fx_symbols:
    fx_raw = download_field_in_chunks(
        fx_symbols,
        field="Close",
        start=START,
        end=END,
        chunk_size=min(len(fx_symbols), 20)
    )

    fx_symbol_to_ccy = {v: k for k, v in FX_TO_EUR.items()}
    fx_rates = fx_raw.rename(columns=fx_symbol_to_ccy).reindex(columns=needed_fx_ccy)
    fx_rates = fx_rates.reindex(prices_native.index).ffill().bfill()
else:
    fx_rates = pd.DataFrame(index=prices_native.index)

# 8) Convertir todo a EUR
prices_eur = prices_native.copy().astype(float)

# 8.1 normalizar unidades (GBp -> GBP)
prices_eur = prices_eur.mul(unit_by_ticker.reindex(prices_eur.columns), axis=1)

# 8.2 aplicar FX contra EUR
for ccy in needed_fx_ccy:
    cols = curr_by_ticker[curr_by_ticker == ccy].index.tolist()
    if cols:
        prices_eur[cols] = prices_eur[cols].mul(fx_rates[ccy], axis=0)

# 9) Versiones con nombres de empresa
ticker_to_name = {ticker: name for name, ticker in NAME_TO_TICKER.items()}

prices_native_named = prices_native.rename(columns=ticker_to_name)
prices_eur_named = prices_eur.rename(columns=ticker_to_name)

returns_native = prices_native.pct_change(fill_method=None)
returns_eur = prices_eur.pct_change(fill_method=None)

returns_native_named = prices_native_named.pct_change(fill_method=None)
returns_eur_named = prices_eur_named.pct_change(fill_method=None)

# 10) Guardar
currency_report.to_csv("/content/stoxx600_currency_report.csv", index=False)

prices_native.to_parquet("/content/stoxx600_prices_native.parquet")
prices_eur.to_parquet("/content/stoxx600_prices_eur.parquet")
returns_native.to_parquet("/content/stoxx600_returns_native.parquet")
returns_eur.to_parquet("/content/stoxx600_returns_eur.parquet")

prices_native_named.to_parquet("/content/stoxx600_prices_native_named.parquet")
prices_eur_named.to_parquet("/content/stoxx600_prices_eur_named.parquet")
returns_native_named.to_parquet("/content/stoxx600_returns_native_named.parquet")
returns_eur_named.to_parquet("/content/stoxx600_returns_eur_named.parquet")

print("\nOK")
print("Guardado en /content/:")
print("- stoxx600_currency_report.csv")
print("- stoxx600_prices_native.parquet")
print("- stoxx600_prices_eur.parquet")
print("- stoxx600_returns_native.parquet")
print("- stoxx600_returns_eur.parquet")
print("- stoxx600_prices_native_named.parquet")
print("- stoxx600_prices_eur_named.parquet")
print("- stoxx600_returns_native_named.parquet")
print("- stoxx600_returns_eur_named.parquet")
```

```python
# ==========================================
# Guardar detalle completo de conversión por activo y fecha
# Reemplaza el bloque anterior por este
# Requiere en memoria:
# - prices_native
# - prices_eur
# - currency_report
# - fx_rates
# - FX_TO_EUR
# ==========================================

import numpy as np
import pandas as pd

# 1) Metadata por ticker
meta = currency_report.copy()

required_cols = {"name", "ticker", "quote_currency_raw", "currency", "unit_factor"}
missing_cols = required_cols - set(meta.columns)
if missing_cols:
    raise RuntimeError(f"Faltan columnas en currency_report: {missing_cols}")

meta = meta.rename(columns={
    "name": "nombre_empresa",
    "quote_currency_raw": "divisa_original",
    "currency": "divisa_original_normalizada",
    "unit_factor": "ajuste_unidad",
})

meta = meta[[
    "nombre_empresa",
    "ticker",
    "divisa_original",
    "divisa_original_normalizada",
    "ajuste_unidad",
]].copy()

meta["ajuste_unidad"] = meta["ajuste_unidad"].astype(float)
meta_by_ticker = meta.set_index("ticker")

# 2) Estandarizar nombres de ejes para que stack/reset_index sea estable
prices_native_std = prices_native.copy()
prices_native_std.index = pd.to_datetime(prices_native_std.index)
prices_native_std.index.name = "fecha"
prices_native_std.columns.name = "ticker"

prices_eur_std = prices_eur.copy()
prices_eur_std.index = pd.to_datetime(prices_eur_std.index)
prices_eur_std.index.name = "fecha"
prices_eur_std.columns.name = "ticker"

# 3) Construir matrices de tasa FX y factor total aplicado
tickers = list(prices_native_std.columns)
dates = prices_native_std.index

fx_market_df = pd.DataFrame(index=dates, columns=tickers, dtype=float)
rate_name_df = pd.DataFrame(index=dates, columns=tickers, dtype=object)
total_factor_df = pd.DataFrame(index=dates, columns=tickers, dtype=float)

for ticker in tickers:
    divisa_norm = meta_by_ticker.loc[ticker, "divisa_original_normalizada"]
    ajuste = float(meta_by_ticker.loc[ticker, "ajuste_unidad"])

    if divisa_norm == "EUR":
        fx_market_df[ticker] = 1.0
        rate_name_df[ticker] = "EUR->EUR"
        total_factor_df[ticker] = ajuste
    else:
        if divisa_norm not in FX_TO_EUR:
            raise RuntimeError(f"No existe par FX definido para la divisa {divisa_norm} del ticker {ticker}")

        nombre_tasa = FX_TO_EUR[divisa_norm]  # p.ej. GBPEUR=X
        if divisa_norm not in fx_rates.columns:
            raise RuntimeError(f"No encuentro la serie FX {divisa_norm} en fx_rates para el ticker {ticker}")

        fx_market_df[ticker] = fx_rates[divisa_norm].astype(float).reindex(dates).ffill().bfill()
        rate_name_df[ticker] = nombre_tasa
        total_factor_df[ticker] = ajuste * fx_market_df[ticker]

# Estandarizar nombres de ejes
fx_market_df.index.name = "fecha"
fx_market_df.columns.name = "ticker"

rate_name_df.index.name = "fecha"
rate_name_df.columns.name = "ticker"

total_factor_df.index.name = "fecha"
total_factor_df.columns.name = "ticker"

# 4) Pasar todo a formato largo
native_long = prices_native_std.stack(dropna=False).rename("precio_original").reset_index()
eur_long = prices_eur_std.stack(dropna=False).rename("precio_eur").reset_index()
fx_market_long = fx_market_df.stack(dropna=False).rename("tasa_fx_mercado").reset_index()
rate_name_long = rate_name_df.stack(dropna=False).rename("nombre_tasa").reset_index()
total_factor_long = total_factor_df.stack(dropna=False).rename("tasa_usada_total_a_eur").reset_index()

# 5) Unir por fecha y ticker
detalle = native_long.merge(eur_long, on=["fecha", "ticker"], how="outer")
detalle = detalle.merge(fx_market_long, on=["fecha", "ticker"], how="left")
detalle = detalle.merge(rate_name_long, on=["fecha", "ticker"], how="left")
detalle = detalle.merge(total_factor_long, on=["fecha", "ticker"], how="left")

# 6) Añadir metadata fija por ticker
detalle = detalle.merge(meta, on="ticker", how="left")

# 7) Añadir divisa final
detalle["divisa_eur"] = "EUR"

# 8) Reordenar columnas
detalle = detalle[[
    "fecha",
    "nombre_empresa",
    "ticker",
    "precio_original",
    "divisa_original",
    "divisa_original_normalizada",
    "ajuste_unidad",
    "tasa_fx_mercado",
    "nombre_tasa",
    "tasa_usada_total_a_eur",
    "precio_eur",
    "divisa_eur",
]].sort_values(["fecha", "ticker"]).reset_index(drop=True)

# 9) Guardar
detalle.to_csv("/content/stoxx600_conversion_detalle.csv", index=False)

print("OK")
print("Guardado en /content/:")
print("- stoxx600_conversion_detalle.csv")
print(detalle.head())
```
